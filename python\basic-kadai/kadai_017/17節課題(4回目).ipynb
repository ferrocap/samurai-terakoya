```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def set_adult_check(self):
        if self.age >= 20:
            print(self.name, "大人である")
        else:
            print(self.name, "大人でない")

humans = [
    Human("侍太郎", 36),
    Human("侍次郎", 30),
    Human("侍三郎", 25),
    Human("侍四郎", 20),
    Human("侍五郎", 17)
]

humans_adult_check = []

# humansリストの各インスタンスでset_adult_check()を呼ぶ
for human in humans:
    humans_adult_check.append(human.set_adult_check())
```
