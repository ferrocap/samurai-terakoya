```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.adult_check = age - 20

    # メソッドを定義する
    def set_age(self, age):
        self.age = age
        self.adult_check = age - 20

    def set_adult_check(self):
        if self.adult_check >= 0:
            return("大人である")
        else:
            return("大人でない")

# クラスの外でHumanのインスタンスを作成する
humans = [
    Human("侍太郎", 36),
    Human("侍次郎", 30),
    Human("侍三郎", 25),
    Human("侍四郎", 20),
    Human("侍五郎", 17)
]

humans_adult_check = []

# humansリストの各インスタンスでset_adult_check()を呼ぶ
for human in humans:
    humans_adult_check.append(human.set_adult_check())

print(humans_adult_check)
```
